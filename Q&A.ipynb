Round 1: The Easy Questions


 Find how many students got a perfect score (100) in Math.

```python
perfect_math = df[df['math_score'] == 100]
count = len(perfect_math)
print(f"Number of Math Geniuses: {count}")
perfect_math
```

Find the number of students who scored less than 40 in Reading.

```python
struggling_readers = df[df['reading_score'] < 40]
count = len(struggling_readers)
print(f"Number of Struggling Readers: {count}")
struggling_readers
```

How many Male vs. Female students are in this dataset?

```python
print(df['gender'].value_counts())
```

What is the mean (average) writing score of the entire class?

```python
avg_writing = df['writing_score'].mean()
print(f"Average Writing Score: {avg_writing:.2f}")
```

Round 2: The Intermediate Questions

Who is better at Math on average? Boys or Girls?

```python
print(df.groupby('gender')['math_score'].mean())
```

 How many students scored above 50 in Math AND Reading AND Writing?

```python
passed_all = df[(df['math_score'] > 50) & (df['reading_score'] > 50) & (df['writing_score'] > 50)]
print(f"Number of All-Rounders: {len(passed_all)}")
```

Show the data for the Top 5 students with the highest average_score.

```python

top_5 = df.sort_values('average_score', ascending=False).head(5)
display(top_5[['gender', 'math_score', 'average_score']])
```

Compare the Average Reading Score for students with "standard" lunch vs. "free/reduced" lunch.

```python

print(df.groupby('lunch')['reading_score'].mean())
```

How many students have parents with a Master's degree?

```python
print(df['parent_education'].value_counts()['master\'s degree'])
```

Round 3: The Hard Questions

Which combination of Parent Education AND Test Prep produces the absolute highest average Math score?

```python
grouped = df.groupby(['parent_education', 'test_prep'])['math_score'].mean()

print(grouped.sort_values(ascending=False))
```

 Find the average Writing Score of students who are in the top 25% for Math.

```python
cutoff = df['math_score'].quantile(0.75)
print(f"Top 25% Math Cutoff Score: {cutoff}")
```

Create a new column called Grade (A, B, C, D, F) based on the average_score.

```python
def assign_grade(score):
    if score >= 90: return 'A'
    elif score >= 80: return 'B'
    elif score >= 70: return 'C'
    elif score >= 60: return 'D'
    else: return 'F'

df['Grade'] = df['average_score'].apply(assign_grade)

print(df['Grade'].value_counts().sort_index())
```

Find students who are incredibly consistent. Their score variance (difference between highest and lowest subject) is less than 10 points.

```python
df['score_range'] = df[['math_score', 'reading_score', 'writing_score']].max(axis=1) - \
                    df[['math_score', 'reading_score', 'writing_score']].min(axis=1)


consistent_students = df[df['score_range'] < 10]
print(f"Number of Consistent Students: {len(consistent_students)}")
print(consistent_students.head())
```

Find students who have "free/reduced" lunch (low income) BUT still managed to score in the Top 10% of the class.

```python
top_10_cutoff = df['average_score'].quantile(0.90)

underdogs = df[(df['lunch'] == 'free/reduced') & (df['average_score'] >= top_10_cutoff)]

print(f"Number of 'Underdog' Students: {len(underdogs)}")

print(underdogs[['lunch', 'average_score']].head())
```

 Which Race/Ethnicity group has the highest percentage of students passing math (>60)?

```python
df['passed_math'] = (df['math_score'] > 60).astype(int)

pass_rates = df.groupby('race')['passed_math'].mean() * 100
print(pass_rates.sort_values(ascending=False))
```

Who has the biggest gap between Reading and Writing?

```python
df['rw_gap'] = abs(df['reading_score'] - df['writing_score'])

max_gap = df['rw_gap'].max()
gap_student = df[df['rw_gap'] == max_gap]

print(f"Largest Gap Found: {max_gap} points")
display(gap_student[['reading_score', 'writing_score', 'rw_gap']])
```

 Calculate the "Advantage" of having a parent with a Master's degree vs. High School for every single subject.

```python
masters_scores = df[df['parent_education'] == "master's degree"][['math_score', 'reading_score', 'writing_score']].mean()
hs_scores = df[df['parent_education'] == 'high school'][['math_score', 'reading_score', 'writing_score']].mean()

advantage = masters_scores - hs_scores
print("--- Master's Degree Advantage (Points Higher) ---")
print(advantage)
```

Downloading the underdogs data 

```python

top_10_cutoff = df['average_score'].quantile(0.90)
underdogs = df[(df['lunch'] == 'free/reduced') & (df['average_score'] >= top_10_cutoff)]

# 2. Select only the columns we care about for the report
report = underdogs[['gender', 'race', 'parent_education', 'math_score', 'reading_score', 'writing_score', 'average_score']]

# 3. Save to CSV
report.to_csv('Scholarship_Candidates_Top_Underdogs.csv', index=False)

print(f"âœ… SUCCESS! Saved {len(report)} students to 'Scholarship_Candidates_Top_Underdogs.csv'")
print("ðŸ“‚ You can now find this file in your folder.")
```
